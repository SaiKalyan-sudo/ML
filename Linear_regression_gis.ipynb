```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sb
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
%matplotlib inline
import seaborn as sns
```

```python
data=pd.read_csv('crime.csv')
data
```

```python
data.size
```

```python
#(Linear Regression)
data.plot(x='YEAR',y='MURDER',style='*',color='c')
plt.title('YEAR vs MURDER',fontdict={'family':'Sans Serif','size':25})
plt.xlabel('YEAR')
plt.ylabel('MURDER')
plt.grid()
plt.show()
```

```python
# we are converting data 2D into 1D data
x=data['YEAR'].values.reshape(-1,1)
y=data['MURDER'].values.reshape(-1,1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
print('Intercept of the curve is:',lr.intercept_)
print('Slope of the curve is:',lr.coef_)
```

```python
y_pred=lr.predict(x_test)
```

```python
pred=lr.predict(x_test)
```

```python
print('Mean Absolute Error:',metrics.mean_absolute_error(y_test, pred))
print('Mean Squared Error:',metrics.mean_squared_error(y_test, pred))
print('Root Mean Squared Error:',np.sqrt(metrics.mean_squared_error(y_test, pred)))
```
